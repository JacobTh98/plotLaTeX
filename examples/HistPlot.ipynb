# Histplot

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

np.random.seed(seed=42)

from plotLaTeX import HistPlot
```

```python
# create random data
n_pts = 1000
data1 = np.random.normal(loc=0, scale=2, size=n_pts)
data2 = np.random.normal(loc=1, scale=2.5, size=n_pts)
```

```python
plt.figure(figsize=(7.5, 3))
plt.hist(data1, bins=20, label="data1", color="C0", alpha=0.5, edgecolor="black")
plt.hist(data2, bins=20, label="data2", color="C1", alpha=0.5, edgecolor="black")
plt.xlabel("x-label")
plt.ylabel("Count")
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

```python
Hplt = HistPlot(bins=10)
```

```python
# add data1
Hplt.add_histdata(data1, name="data1")
```

```python
# add data2
Hplt.add_histdata(data2, name="data2")
```

```python
Hplt.add_axis_labels(xlabel="x-label", ylabel="Count")
```

```python
Hplt.export()
```

```python
# plot the data using seaborn
sns.histplot(data=Hplt.data, bins=Hplt.bins)
```
